```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("./data/preprocessing_phones.csv")
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,0:10].values
y=dataset.iloc[:,10].values
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
```

```python
#Lineer Regresyon ile
lr = LinearRegression()
lr.fit(X_train,y_train)
# 
y_predict = lr.predict(X_test) # üretilen modeli (lr) test vverisi ile deneyelim
y_predict  #tahmin ettiği değerleri görelim

```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
lineer_rmse=np.sqrt(mean_squared_error(y_test,y_predict))
lineer_r2=r2_score(y_test,y_predict)

print(' root mean squared error: ',lineer_rmse)
print('Coefficient of determination R^2: ',lineer_r2)
```

```python
import joblib
joblib.dump(lr,"./Models/phones_lineer.pkl")
```
