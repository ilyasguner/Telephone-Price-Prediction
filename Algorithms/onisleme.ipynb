```python
columns = ['cpu', 'dahili_hafiza', 'ekran_boyutu',
       'kamera_cozunurlugu','mobil_baglanti_hizi',"pil_gucu",'ram_kapasitesi','on_kamera_cozunurluk',
        'isletim_sistemi','brands',"price"]


def distplot(nrows, ncols, columns):

    rows=nrows
    cols=ncols

    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(18, 12))

    columns = columns
    index=0

    for i in range(rows):
        for j in range(cols):
            sns.distplot(data[columns[index]], ax=ax[i][j], bins=20)
            index+=1

        
distplot(3, 3, columns)
```
